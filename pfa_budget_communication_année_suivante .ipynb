```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import sklearn 
import seaborn as sns 
```

```python
df = pd.read_csv("Downloads/dataset - Sheet1 (3).csv")
```

```python
df 
```

```python
df2 = df.drop(["Entreprise","Prix","clientele","part_Marche"],axis="columns")
```

```python
df2["pourcentage"] = (df2["Budget annee suivante"].astype(float)/df2["Chiffre affaires par an"].astype(float))*100
```

```python
df2["pourcentage_1"] =  (df2["Budget de la communication par an"].astype(float)/df2["Chiffre affaires par an"].astype(float))*100
```

```python
df2 
```

```python
df2 = df2[(df2["pourcentage"]<=40) & (df2["pourcentage"]>=10)]  
df2 = df2[(df2["pourcentage_1"]<=40) & (df2["pourcentage_1"]>=10)] 

df2 
```

```python
df2 = df2.drop(["pourcentage","Chiffre affaires par an"],axis="columns") 
```

```python
df2["la categorie du produit"]= df2["la categorie du produit"].map({"produit consommable":"0","produit fini":"1","Services informatiques":"2","Packaging":"3","Composants automobiles":"4","Electronique":"5","service":"6"})
```

```python
df2 
```

```python
dummies = pd.get_dummies(df2["produit"])  
type(dummies)
```

```python
df_merge   =  pd.concat([df2,dummies],axis="columns" )  
```

```python
df_merge 
```

```python
df_merge.isna().sum()
```

```python
df_merge = df_merge.drop(["produit"],axis="columns")
```

```python
x_train = df_merge.drop(["Budget annee suivante"],axis="columns")
y_train = df_merge["Budget annee suivante"] 
y_train = y_train.astype(float)
```

```python
plt.plot(x_train["Budget de la communication par an"],y_train) 
plt.xlabel("Budget de la communication par an") 
plt.ylabel("Budget ann√©e suivante") 
```

```python
y_train 
```

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x_train,y_train,test_size = 0.2,random_state=42)
```

```python
y_test 
```

```python
from sklearn import linear_model 
from sklearn.model_selection import GridSearchCV    
model = linear_model.LinearRegression()

parameters = {'fit_intercept':[True,False], 'copy_X':[True, False]}
grid = GridSearchCV(model,parameters, cv=None) 
grid.fit(x_train, y_train)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_train)
```

```python
model.score(x_train,y_train)
```

```python
grid.predict(X_test)
```

```python
grid.score(X_test,y_test)
```
