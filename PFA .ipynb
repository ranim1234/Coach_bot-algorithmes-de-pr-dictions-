```python
y_train = df4["clientele"]
```

```python
x_train = df4.drop(["clientele"],axis="columns")
```

```python
x_train.isna().sum() 
```

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x_train,y_train,test_size = 0.2,random_state=42)
```

```python
from sklearn import svm 
from sklearn import naive_bayes 
from sklearn import ensemble 
from sklearn import tree 
from sklearn import linear_model 
```

```python
model = naive_bayes.MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test) 
```

```python
model_1 = svm.SVC()
```

```python
model_1.fit(x_train,y_train)
```

```python
model_1.score(X_test,y_test)
```

```python
model_params = {
    "svm":{
        "model":svm.SVC(gamma="auto"),
        "params":{
            "C":[1,5,10],
            "kernel":["rbf","linear"]
        }
    } , 
   "randomforest":{
       "model":ensemble.RandomForestClassifier() ,
       "params":{
           "n_estimators":[1,5,10]
       }
   }
        
        
    ,
 "tree":{
   "model":tree.DecisionTreeClassifier()  , 
     "params":{
                     'criterion': ['gini','entropy']

     }
 },
 
 "naivebayes":{
     "model":naive_bayes.GaussianNB() ,
     "params":{}
     
 },

    'logistic_regression' : {
        'model': linear_model.LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    }
 

    
    
    
    
    }
```

```python
from sklearn.model_selection import GridSearchCV    
```

```python
result = []
for key , value in model_params.items() : 
        model = GridSearchCV(value["model"],value["params"],cv=3,return_train_score = False ) 
        model.fit(x_train,y_train) 
        model_score =  model.score(X_test,y_test)
        result.append({
            "model": key ,
            "best_score" : model.best_score_ , 
            "best params " : model.best_params_ , 
            "test_score": model_score
        })
        
```
